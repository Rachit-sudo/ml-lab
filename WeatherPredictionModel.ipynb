```python
predictors = ["precip","temp_max","temp_min"]
```

```python
train = core_weather.loc[:"2020-12-31"] #split the training set
```

```python
test = core_weather.loc["2021-01-01":] #split the testing set
```

```python
#use previous data to predict the future observation or we get a horrible data
reg.fit(train[predictors],train["target"])
```

```python
predictions = reg.predict(test[predictors])
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
#create a function to make predicition
def create_predictions(predictors,core_weather,reg):
    train = core_weather.loc[:"2020-12-31"]
    test = core_weather.loc["2021-01-01":]
    reg.fit(train[predictors],train["target"])
    predictions = reg.predict(test[predictors])
    error = mean_absolute_error(test["target"],predictions)
    combined = pd.concat([test["target"],pd.Series(predictions, index=test.index)], axis=1) #combine the actuals to see the actual value in the prediciton side by side and concatenate, first we need to convert them
    combined.columns = ["actual","predictions"]
    return error, combined
#instead of copy and pasting or iterating we can run our function directly
```

```python
#Adding in Rolling Means
#create more predictors
#average temperature
core_weather["month_max"] = core_weather["temp_max"].rolling(30).mean()
```

```python
core_weather["max_min"] = core_weather["temp_max"] / core_weather["temp_min"]
#what the temperature will actually do
```

```python
core_weather["precip"] = core_weather["precip"].fillna(0)
core_weather["temp_max"] = core_weather["temp_max"].fillna(0)
core_weather["temp_min"] = core_weather["temp_min"].fillna(0)
core_weather["month_max"] = core_weather["month_max"].fillna(0)
core_weather["month_day_max"] = core_weather["month_day_max"].fillna(0)
core_weather["max_min"] = core_weather["max_min"].fillna(0)
predictors = ["precip","temp_max","temp_min","month_max","month_day_max","max_min"]
```

```python
#remove first thirty rows which have the most NaN values
core_weather = core_weather.iloc[35:,:].copy()
```
